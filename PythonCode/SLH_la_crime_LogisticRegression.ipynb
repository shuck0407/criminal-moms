```python
#Import dependencies
import pandas as pd
import sqlite3
import sys
```

```python
#Assign database file to a variable
db_file = "Data/crimedata.db"

#Connect to database file
conn = sqlite3.connect(db_file)

```

```python
#Query the la crime data table and put into a pandas dataframe
crime_df = pd.read_sql_query("select * from la_crime_1718;", conn)
crime_df.head()
```

```python
#Query the la weather history data table and put into a pandas dataframe
weather_df = pd.read_sql_query("select * from LA_weather_history_2016_to_2018;", conn)
weather_df.head()
```

```python
#Merge the weather history with the crime data

#First, strip the first two digits of the time occured column so that we can match it up with the weather

crime_df['hour'] = crime_df['TimeOccurred'].str.slice(0,2)
crime_df.head()
```

```python
new_df = pd.merge(crime_df, weather_df, left_on=['DateOccurred', 'hour'], right_on=['dt_iso', 'hour'])
new_df.head()
```

```python
#Drop columns from the dataframe that will not be needed in our ML models

model_df = new_df.drop(['DRNumber', 'DateReported', 'DateOccurred', 'AreaName', 'CrimeCodeDescription', 'MOCodes',
                       'PremiseDescription', 'weather_description', 'weather_main', 'rain_1h', 'rain_3h'],
                     axis=1)
model_df.head()
```

```python
model_df = model_df.drop(['WeaponDescription', 'VictimSex', 'VictimDescent', 'StatusCode', 'StatusDescription',
                         'Address', 'CrossStreet', 'CrimeCode1', 'CrimeCode2', 'CrimeCode3', 'CrimeCode4','dt_iso'], axis=1)
model_df.head()
```

```python
model_small_df = model_df.loc[model_df['CrimeCode'].isin(['420', '210', '230', '626', '740',
 '310', '510', '440', '330', '624'])]
model_small_df.head()
```

```python
#COnvert the datatype of each column to numeric for machine learning
model_small_df['TimeOccurred'] = model_small_df['TimeOccurred'].apply(pd.to_numeric, errors='coerce')
model_small_df['AreaID'] = model_small_df['AreaID'].apply(pd.to_numeric, errors='coerce')
model_small_df['CrimeCode'] = model_small_df['CrimeCode'].apply(pd.to_numeric, errors='coerce')
model_small_df['ReportingDistrict'] = model_small_df['ReportingDistrict'].apply(pd.to_numeric, errors='coerce')
model_small_df['VictimAge'] = model_small_df['VictimAge'].apply(pd.to_numeric, errors='coerce')
model_small_df['PremiseCode'] = model_small_df['PremiseCode'].apply(pd.to_numeric, errors='coerce')
model_small_df['WeaponUsedCode'] = model_small_df['WeaponUsedCode'].apply(pd.to_numeric, errors='coerce')
model_small_df['Location'] = model_small_df['Location'].apply(pd.to_numeric, errors='coerce')
model_small_df['hour'] = model_small_df['hour'].apply(pd.to_numeric, errors='coerce')
model_small_df['temp'] = model_small_df['temp'].apply(pd.to_numeric, errors='coerce')
model_small_df['temp_min'] = model_small_df['temp_min'].apply(pd.to_numeric, errors='coerce')
model_small_df['temp_max'] = model_small_df['temp_max'].apply(pd.to_numeric, errors='coerce')
model_small_df['pressure'] = model_small_df['pressure'].apply(pd.to_numeric, errors='coerce')
model_small_df['humidity'] = model_small_df['humidity'].apply(pd.to_numeric, errors='coerce')
model_small_df['wind_speed'] = model_small_df['wind_speed'].apply(pd.to_numeric, errors='coerce')
model_small_df['wind_deg'] = model_small_df['wind_deg'].apply(pd.to_numeric, errors='coerce')
model_small_df['clouds_all'] = model_small_df['clouds_all'].apply(pd.to_numeric, errors='coerce')
model_small_df['weather_id'] = model_small_df['weather_id'].apply(pd.to_numeric, errors='coerce')


print(model_small_df.dtypes)
model_small_df
```

```python
model_small_df = model_small_df.fillna(0)
model_small_df.head()
```

```python
model_small_df.reset_index()
```

```python
# Assign X (data) and y (target) to put into model
X = model_small_df.drop("CrimeCode", axis=1)
y = model_small_df["CrimeCode"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```
